```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def doOperationWithImage(img, const, operation="None"):
    if operation == "add":
        return cv2.add(img, const)
    elif operation == "sub":
        return cv2.subtract(img, const)
    elif operation == "mul":
        return cv2.multiply(img, const)
    elif operation == "div":
        return cv2.divide(img, const)
    else:
        raise ValueError("Invalid operation")
```

```python
imagePath = "images/week3/camel.png"
image = cv2.imread(imagePath, cv2.IMREAD_COLOR)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# -- DO ADD OPERATION --
constLstForAddSub = [-100, -50, 50, 100]
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle("Add operation")
for i, ax in enumerate(axs.flat):
    ax.imshow(doOperationWithImage(image, const=constLstForAddSub[i], operation="add"))
    ax.set_title(f"Add {constLstForAddSub[i]}")
plt.show()

# -- DO SUB OPERATION --
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle("Subtract operation")
for i, ax in enumerate(axs.flat):
    ax.imshow(doOperationWithImage(image, const=constLstForAddSub[i], operation="sub"))   
    ax.set_title(f"Subtract {constLstForAddSub[i]}")
plt.show()

constLstForMulDiv = [0.1, 0.5, 1, 5]

# -- DO MUL OPERATION --
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle("Multiply operation")
for i, ax in enumerate(axs.flat):
    ax.imshow(doOperationWithImage(image, const=constLstForMulDiv[i], operation="mul"))
    ax.set_title(f"Multiply {constLstForMulDiv[i]}")
plt.show()

# -- DO DIV OPERATION --
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle("Divide operation")
for i, ax in enumerate(axs.flat):
    ax.imshow(doOperationWithImage(image, const=constLstForMulDiv[i], operation="div"))
    ax.set_title(f"Divide {constLstForMulDiv[i]}")
plt.show()
```

```python
# Enhance Image Quality: substract 100, multiply 5
enhancedImage = doOperationWithImage(image, const=100, operation="sub")
enhancedImage = doOperationWithImage(enhancedImage, const=5, operation="mul")
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle("Enhanced Image")
axs[0].imshow(enhancedImage)
axs[1].hist(enhancedImage.ravel(), 256, [0, 256])
plt.ylim([0, 5000])
plt.tight_layout()
plt.xlabel("Pixel intensity")
plt.ylabel("Frequency (count)")
plt.show()
```
