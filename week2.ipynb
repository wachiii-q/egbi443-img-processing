```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
sameHisImagePath1 = "images/week2_histogram/pic_1.png"
sameHisImagePath2 = "images/week2_histogram/pic_2.png"
sameHisImagePath3 = "images/week2_histogram/pic_3.png"

# plot the histogram
def plotHistogram(imagePath): 
    img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE) # what imread do is to read the image in BGR format
    hist, bins = np.histogram(img.flatten(), 256, [0, 256]) # what histogram do is to calculate the histogram of the image what it returns is the histogram and the bins
    plt.hist(img.flatten(), 256, [0, 256], color='b')
    plt.xlim([0, 256])  
    plt.show()
    
plotHistogram(sameHisImagePath1)
plotHistogram(sameHisImagePath2)
plotHistogram(sameHisImagePath3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Assignment 2: plot the histogram of low contrast image and high contrast image then compare them
imgHighContrastPath = "images/week2_histogram/mount_high_contrast.png"
imgLowContrastPath = "images/week2_histogram/mount_low_contrast.png"


def plotHistogram(imagePath):
    img = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)
    hist, bins = np.histogram(img.flatten(), 256, [0, 256])
    plt.hist(img.flatten(), 256, [0, 256], color='b')
    plt.xlim([0, 256])  
    plt.show()


# plot the histogram of high contrast image
plotHistogram(imgHighContrastPath)  # what x-axis and y-axis represent? x-axis is the intensity of the pixel, y-axis is the number of pixels with that intensity

# plot the histogram of low contrast image
plotHistogram(imgLowContrastPath)
```
