```python
import cv2
import matplotlib.pyplot as plt
from icecream import icecream as log

input = cv2.imread('/Users/wachiii/Desktop/image_processing_uni/images/dango.jpg') 
print("shape of image ", input.shape)

# use matplotlib to display the image
plt.imshow(cv2.cvtColor(input, cv2.COLOR_BGR2RGB))
```

```python
import cv2
import matplotlib.pyplot as plt
from icecream import icecream as log

# -- split into 3  images; r g b
b, g, r = cv2.split(input) # what split does is it splits the image into its three channels, blue, green, and red. if we plot each of these channels, we will see the image in its three channels.

# -- 2nd plot: show the images
plt.figure(figsize=(20, 20))
# ---- plot red channel
plt.subplot(131)        
plt.imshow(r, cmap='gray')
plt.title('Red Channel')
# ---- plot green channel
plt.subplot(132)
plt.imshow(g, cmap='gray')
plt.title('Green Channel')
# ---- plot blue channel
plt.subplot(133)
plt.imshow(b, cmap='gray')
plt.title('Blue Channel')
plt.show()
```
